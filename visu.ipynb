```python
import xarray as xr
import matplotlib.pyplot as pl
```

```python
ds = xr.open_dataset("tmp/data.nc")
da = ds['__xarray_dataarray_variable__'].rename('rho_toa')
da.coords['model']
da.coords['rho_s']
```

```python
lin_sulfate_hr30_80_bc_20_550nm = da.sel(model="sulfate_hr30_only_550nm")*0.8 + da.sel(model="bc_only_550nm")*0.2
xval1 = 0
xval2 = 1
yval1 = 0
yval2 = 1

fig, axs = pl.subplots(2, 2, figsize=(12,12))

axs[0, 0].set_title('100% Black Carbon')
axs[0, 0].set_xlabel('rho_s')
axs[0, 0].set_ylabel('rho_toa')
axs[0, 0].set_xlim(xval1,xval2)
axs[0, 0].set_ylim(yval1,yval2)
axs[0, 0].plot(da.coords['rho_s'], da.sel(model="bc_only_550nm", tau=0), 'b', label="tau=0.0")
axs[0, 0].plot(da.coords['rho_s'], da.sel(model="bc_only_550nm", tau=0.4), 'g', label="tau=0.4")
axs[0, 0].plot(da.coords['rho_s'], da.sel(model="bc_only_550nm", tau=0.8), 'r', label="tau=0.8")
axs[0, 0].legend()

axs[0, 1].set_title('100% Sulfate (HR=30%)')
axs[0, 1].set_xlabel('rho_s')
axs[0, 1].set_ylabel('rho_toa')
axs[0, 1].set_xlim(xval1,xval2)
axs[0, 1].set_ylim(yval1,yval2)
axs[0, 1].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_only_550nm", tau=0), 'b', label="tau=0.0")
axs[0, 1].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_only_550nm", tau=0.4), 'g', label="tau=0.4")
axs[0, 1].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_only_550nm", tau=0.8), 'r', label="tau=0.8")
axs[0, 1].legend()

axs[1, 0].set_title('SOS_ABS multimode: 80% Sulfate, 20% Black Carbon')
axs[1, 0].set_xlabel('rho_s')
axs[1, 0].set_ylabel('rho_toa')
axs[1, 0].set_xlim(xval1,xval2)
axs[1, 0].set_ylim(yval1,yval2)
axs[1, 0].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_80_bc_20_550nm", tau=0), 'b', label="tau=0.0")
axs[1, 0].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_80_bc_20_550nm", tau=0.4), 'g', label="tau=0.4")
axs[1, 0].plot(da.coords['rho_s'], da.sel(model="sulfate_hr30_80_bc_20_550nm", tau=0.8), 'r', label="tau=0.8")
axs[1, 0].legend()

axs[1, 1].set_title('Monomod. combination: 80% Sulfate, 20% Black Carbon')
axs[1, 1].set_xlabel('rho_s')
axs[1, 1].set_ylabel('rho_toa')
axs[1, 1].set_xlim(xval1,xval2)
axs[1, 1].set_ylim(yval1,yval2)
axs[1, 1].plot(da.coords['rho_s'], lin_sulfate_hr30_80_bc_20_550nm.sel(tau=0), 'b', label="tau=0.0")
axs[1, 1].plot(da.coords['rho_s'], lin_sulfate_hr30_80_bc_20_550nm.sel(tau=0.4), 'g', label="tau=0.4")
axs[1, 1].plot(da.coords['rho_s'], lin_sulfate_hr30_80_bc_20_550nm.sel(tau=0.8), 'r', label="tau=0.8")
axs[1, 1].legend()

pl.savefig("sulfate_hr30_80_bc_20_550nm.png")
```
